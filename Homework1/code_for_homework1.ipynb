```python
import numpy as np
import math as mt
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.sin(np.pi*x)*np.sin(np.pi*x)
```

```python
x = np.arange(-2,2)
```

```python
u_exata = f(x)
```

```python
plt.figure(figsize=(18,10))
plt.plot(x, u_exata, label='u_exata')
plt.xlabel('x')
plt.ylabel('u_exata')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.linalg.norm(np.abs(u_exata), np.inf)
```
